and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
openai = OpenAI()

stream = openai.chat.completions.create(model=MODEL_GPT, messages=[{"role": "user", "content":question}], stream=True)

response = ""
for chunk in stream:
    print(chunk.choices[0].delta.content or "", end="", flush=True)
```

```python
# Get Llama 3.2 to answer
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
streamo = ollama.chat.completions.create(model=MODEL_LLAMA, messages=[{"role": "user", "content":question}], stream=True)

response = ""
for chunk in streamo:
    print(chunk.choices[0].delta.content or "", end="", flush=True)
```
